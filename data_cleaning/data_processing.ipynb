```python
def clean_price(x):
    if not isinstance(x, str):
        return np.nan

    x = x.strip()

    # If bad text, return NaN
    if x.lower() in ["price", "contact", "on request", "-", "nan", ""]:
        return np.nan

    parts = x.split()

    # extract number (remove ₹)
    num = re.findall(r"[\d\.]+", x)
    if not num:
        return np.nan
    num = float(num[0])

    # detect Crore or Lac
    if "lac" in x.lower():
        return round(num / 100, 2)   # convert Lac → Crore
    else:
        return round(num, 2)         # Cr stays Cr
```

```python
df['price'] = df['price'].apply(clean_price)
```

```python
def parse_loc(lst):
    if not isinstance(lst, list):
        return [None, None, None]
    
    lst = [x.strip() for x in lst]  # remove spaces
    
    if len(lst) == 3:
        return [lst[0], lst[1], lst[2]]
    elif len(lst) == 2:
        return [None, lst[0], lst[1]]
    elif len(lst) == 1:
        return [None, lst[0], None]
    else:
        return [None, None, None]
```

```python
df[['sub_location', 'main_location', 'city']] = df['loc_list'].apply(
    lambda x: pd.Series(parse_loc(x))
)
```
